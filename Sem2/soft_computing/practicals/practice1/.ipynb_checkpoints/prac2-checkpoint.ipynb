```python
inData =[[-1,-1],[-1,1],[1,-1],[1,1]]

weights =[0,0]
bias=0
```

```python
#and gate
weightHistory1 =[weights[0]]
weightHistory2=[weights[1]]
expectedOutputs =[-1,-1,-1,1]

for index,data in enumerate(inData):
    weights[0] +=(expectedOutputs[index]*data[0])
    weights[1] +=(expectedOutputs[index]*data[1])
    bias += expectedOutputs[index]
    weightHistory1.append(weights[0])
    weightHistory2.append(weights[1])

print(weights, bias)
        
```

```python
# check predictions 
for data in inData:
    summation = sum([i*w for i,w in zip(data, weights)],bias)
    print("Input:", data, -1 if summation<1 else 1)
```

```python
# or gate
expectedOutputs = [-1,1,1,1]
weights =[0,0]
bias=0

for idx,data in enumerate(inData):
    weights[0]+= expectedOutputs[idx]*data[0]
    weights[1]+=expectedOutputs[idx]*data[1]
    bias +=expectedOutputs[idx]

# predictions 
for data in inData:
    print("Input:", data, -1 if sum([i*w for i,w in zip(data, weights)],bias)<1 else 1)
```

```python
#plot the weight history
import matplotlib.pyplot as plt

plt.plot([1,2,3,4,5], weightHistory1,marker="o", label="weight 1")
plt.plot([1,2,3,4,5], weightHistory2, marker="s", label="weight 2")
plt.xlabel("Training steps")
plt.ylabel("Weights value")
plt.xticks(ticks = range(len(weightHistory1)), labels = [f"Step {step}" for step in range(1,len(weightHistory1)+1)])
plt.grid()
plt.legend()
plt.title("Diamond diamond nice nice")
```
