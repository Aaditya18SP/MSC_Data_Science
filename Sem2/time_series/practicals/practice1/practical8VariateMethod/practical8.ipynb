```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../../data/Symphony-Data.csv")
data.head()
```

```python
priceDiff = data["PRICE"].diff()
priceDiff
```

```python
meanPriceDiff = priceDiff.mean()
varPriceDiff = priceDiff.var()
print("Mean", meanPriceDiff)
print("Variance of differences", varPriceDiff)
```

```python
#because the errors are independently and identically distributed
#hence , var(e-e_prev) = var(e)+var(e_prev) = v^2+v^2 = 2v^2 
varRandomComp = varPriceDiff/2
varRandomComp
```
