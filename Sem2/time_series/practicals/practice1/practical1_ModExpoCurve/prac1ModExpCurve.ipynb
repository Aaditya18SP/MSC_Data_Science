# Theory - Fitting a modified exponential curve
- modiried exponential curve formula  - ae^(bx) +c
- 'c' is the modification that says that the value was never zero. Eg: atmospheric C02 levels were never 0 hence a small y-intercept
- steps:
  -  Data must be in format - (date, value)
  -  plot the data
  -  derive initial estimates for a, b,c 
  -  fit the curve using scipy.optimize.curve_fit
  -  get the optimized parameters
  -  plot the covariance matrix between the parameters
  -  plot the data and the curve

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = pd.read_csv("./prac1/data/co2-ppm-daily.csv")
data.head()
```

```python
data.dtypes
```

```python
data["date"] = pd.to_datetime(data["date"])
data = data.sort_values(by="date", ascending=True)
data.head()
```

```python
data.dtypes
```

```python
x = data["date"]
y = data["value"]
```

```python
#convert to numerical form to feed it into the curve_fit function
xNum = data["date"] - data["date"].min()
print(xNum)
print()
xNum = xNum.dt.days
xNum
```

```python
#plot the data
plt.plot(x,y, label="C02 emission")
plt.xlabel("Date")
plt.ylabel("C02 levels(in MT)")
plt.title("C02 emission over time")
plt.grid()
plt.legend()
```

```python
def modifiedExponentialCurve(t,a,b,c):
    return a*np.exp(b*t)+c

if len(x)%2==0:
    middleIdx = len(x)//2 -1
else:
    middleIdx = len(x)//2
    
initialA = y[0]
initialC= y[0]
initialB = 1/pd.Timestamp(x[middleIdx]).year

popt, pcov = curve_fit(modifiedExponentialCurve, xNum, y, p0=[initialA, initialB, initialC])

print("optimized paramters are :",popt)
a,b,c =popt
```

```python
ypred = modifiedExponentialCurve(xNum, a,b,c)
ypred
```

```python
plt.plot(x,y, label="Original", color="blue")
plt.plot(x, ypred, label="Curve", color="red")
plt.xlabel("Year")
plt.ylabel("Emission over time(in MT)")
plt.title("Curve fit")
plt.legend()
plt.grid()
```

```python
# covariance grid
cax = plt.imshow(pcov, cmap="YlGn")
plt.grid(which="minor", color="b")
for i in range(pcov.shape[0]):
    for j in range(pcov.shape[1]):
        plt.text(j,i, f"{pcov[i,j]:.2f}")
plt.xticks([0,1,2])
plt.yticks([0,1,2])
plt.colorbar(cax)
```
