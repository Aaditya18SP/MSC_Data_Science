```python
#prac 3
#data cleaning and preparation

import pandas as pd
import seaborn as sns


data = sns.load_dataset("titanic")

print(data.head())

# check for missing values
print(data.isnull().sum())

# handle missing values
data["age"] = data["age"].fillna(data["age"].median())

data["embarked"] = data["embarked"].fillna(data["embarked"].mode()[0])
data = data.drop(columns="deck")

data["embark_town"] = data["embark_town"].fillna(data["embark_town"].mode()[0])
print()
print()
print(data.isnull().sum())


```

```python
#prac4
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler
import matplotlib.pyplot as plt
#handle categorical variables

scaler = StandardScaler()
labelenc = LabelEncoder()
#data[["age","fare"]] = scaler.fit_transform(data[["age","fare"]])
data["embarked"] = labelenc.fit_transform(data["embarked"])
data["sex"] = labelenc.fit_transform(data["sex"])

data["who"] = data["who"].apply(lambda x: 1 if x=="man" else 0)
print(data.head())

print()
print(data.columns)
#feature engineering
data["family_size"] = data.loc[:,"parch"]+ data.loc[:,"sibsp"]+1

#remove duplicates
data = data.drop_duplicates()

#detect and handle and cap outliers

#cap fare to 99th percentile
fare_cap = data["fare"].quantile(0.75)
data["fare"] = data.loc[:,"fare"].apply(lambda x: fare_cap if x > fare_cap else x)


print(data.shape)
data2 = data.select_dtypes(include="number")
fig,ax = plt.subplots(ncols=3, nrows=5,figsize=(10,15) )
ax=  ax.flatten()
idx =0
for k,v in data2.items():
  ax[idx].set_ylabel(k)

  ax[idx].boxplot(v)
  idx+=1
plt.show()

#data = data.drop(columns="family_size")

# Normalize data
scaler = RobustScaler()
normalizer = StandardScaler()
data["age"] = scaler.fit_transform(data[["age"]])
data["fare"] = normalizer.fit_transform(data[["fare"]])

print(data.head())
```
