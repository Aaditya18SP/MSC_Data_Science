```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("../given/practical6/METABRIC_RNA_Mutation.csv")
data.head()
```

```python
data=data.dropna()
x =data.drop("death_from_cancer", axis=1)
y =data["death_from_cancer"]
```

```python
x = pd.get_dummies(x)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
x_tr, x_test, y_tr, y_test = train_test_split(x, y, test_size=0.2, random_state=200)
```

```python
# LogisticRegression
```

```python
model = LogisticRegression().fit(x_tr, y_tr)
predTrain = model.predict(x_tr)
predTest = model.predict(x_test)

print(confusion_matrix(y_tr, predTrain))
print(classification_report(y_tr, predTrain))
print()
print(confusion_matrix(y_test, predTest))
print(classification_report(y_test, predTest))
```

```python
model = DecisionTreeClassifier().fit(x_tr, y_tr)
predTrain = model.predict(x_tr)
predTest= model.predict(x_test)
print("Train")
print(confusion_matrix(y_tr, predTrain))
print()
print(classification_report(y_tr, predTrain))
print()
print()
print("Test")
print(confusion_matrix(y_test, predTest))
print()
print(classification_report(y_test, predTest))
```

```python
# find k for Select K Best 
param={"kbest__k":[10,20,30, 40, 50, 100, 200, 400, 1000,2000, 4000, 5000]}
model = GridSearchCV(Pipeline([("kbest", SelectKBest(f_classif)), ("classify", RandomForestClassifier())]),param_grid=param, cv=10)
model.fit(x,y)
model.best_params_
```
