```python
import pandas as pd
```

```python
data = pd.read_csv("/content/youtube.csv")

data.dtypes
```

```python
def rankMapper(ele):
  if "st" in ele:
    return ele.removesuffix("st")
  elif "nd" in ele:
    return ele.removesuffix("nd")
  elif "rd" in ele:
    return ele.removesuffix("rd")
  elif "th" in ele:
    return ele.removesuffix("th")
  else:
    return ele

def uMapper(row):
  if row=="nan":
    return 0
  else:
    return row
def sMapper(ele):
  if "M" in ele:
    return ele.removesuffix("M")
  elif "K" in ele:
    return ele.removesuffix("K")
  else:
    return ele

def vMapper(ele):
  if ele == "--":
    return 0
  else:
    return ele
#data["Rank"] = data["Rank"].map(rankMapper).astype("float")

#data["Uploads"] = data["Uploads"].map(lambda x : "".join(str(x).split(","))).map(uMapper).astype("int")
#data["Subscriptions"] = data["Subscriptions"].map(lambda x : "".join(x.split(","))).map(sMapper).astype("float")

data["Views"] = data["Views"].map(lambda x: "".join(x.split(","))).map(vMapper).astype("int")
data.dtypes
```

```python
# drop rows
data = data.drop(index=[x for x in range(2,44)])

#drop columns
data = data.drop(columns=["Grade"])
data.head()
```

```python
#group by and sorting
data.sort_values("Subscriptions", ascending=False).head()

data.groupby("Ch_name").agg({"Rank":"max", "Uploads":"sum", "Subscriptions":"sum"}).sort_values(by=["Rank","Uploads", "Subscriptions"], ascending=[False,False,True]).head()
```
