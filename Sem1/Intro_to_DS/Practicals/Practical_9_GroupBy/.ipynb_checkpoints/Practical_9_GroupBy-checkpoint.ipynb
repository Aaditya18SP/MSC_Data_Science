```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('bank_csv.csv')
data.head()
```

```python
data.shape
```

```python
# group by job status
#data["job"].unique()
jb=data.groupby(by="job")["balance"].sum()
jb.reset_index().sort_values(by="balance", ascending =False)
```

```python
# the mean balance by month
mb=data.groupby(by="month")["balance"].mean()
mb.reset_index().sort_values(by="balance", ascending =False) 
```

```python
# the average age grouped by job and martial status and sort them accordingly
jmb = data.groupby(by=["job","marital"])["age"].mean()
jmb.reset_index().sort_values(by="age",ascending=True)
```

```python
# mean age of people with house loan
with_housing_loan = data[(data["housing"]=="yes") & (data["loan"]=="yes")]
housing_loan =with_housing_loan.groupby(by=["housing","loan","job"])["age"].mean()
housing_loan =housing_loan.reset_index().sort_values(by="age")
housing_loan.drop(columns=["housing","loan"])
```

```python
# mean balance of people with house loan
hl_balance= with_housing_loan.groupby(by=["housing","loan","job"])["balance"].mean()
h1_balance= hl_balance.reset_index().sort_values(by="balance", ascending=False)
h1_balance.drop(columns=["housing","loan"])
```
