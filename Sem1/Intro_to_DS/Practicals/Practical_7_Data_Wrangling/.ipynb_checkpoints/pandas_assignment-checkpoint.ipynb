```python
import pandas as pd
import numpy as np
```

```python
# 1. import the dataset
df = pd.read_csv('youtube.csv')
```

```python
df.head()
```

#### 2. print the data types

```python
#converting columns to numeric type
df["Subscriptions"]=df["Subscriptions"].str.extract(r'(\d+)')
df["Subscriptions"]=pd.to_numeric(df["Subscriptions"])
df["Uploads"]=df["Uploads"].str.replace(',','')
df["Uploads"]=pd.to_numeric(df["Uploads"])
df["Views"]=df["Views"].str.replace(',','')
df["Views"]=df["Views"].str.replace('--','')
df["Rank"]=df["Rank"].str.extract(r'(\d+)')
df["Views"]=pd.to_numeric(df["Views"])
```

#### 5. print details of the channels T-Series,SAB TV and Zee TV

```python
df[(df["Ch_name"]=="T-Series") | (df["Ch_name"]=="SAB TV") | (df["Ch_name"]=="Zee TV") ]
```

#### 6. print the channels where subscription is between 50 and 100 million

```python
df[(df["Subscriptions"] >= 50) & (df["Subscriptions"]<=100)]

```

#### 7.What are the top 10 youtube channels as ordered by ‘subscriptions in million’?**

```python
df.sort_values("Subscriptions", ascending=False).head(10)
```

#### 8. Generate a table which only contains data for channels which have less than 5000 video uploads. Sort this table by Rank.**

```python
(df[df["Uploads"]<5000]).sort_values("Rank",ascending = True)
```

#### 9.  Print the average number of subcribers per grade type

```python
df1= df.groupby(["Grade"])["Subscriptions"].mean()
df1
```

#### 10. What are the top 5 channels which belong to grade A++ based on Video_views?

```python
df[df["Grade"]=="A++"].sort_values("Views",ascending=False).head(5)
```
